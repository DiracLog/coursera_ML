```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from numpy import exp
from sklearn.metrics import log_loss
import numpy as np
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
%matplotlib inline

```

```python
path_to_data = "C:/study/data/"
```

```python
df = pd.read_csv(path_to_data + "gbm-data.csv")
```

```python
df.head(20)
#print(len(df['Activity']))
#df.tail()
```

```python
y= df['Activity'].to_numpy()
```

```python
X = df[df.columns[1:]].to_numpy()
```

```python
def sigmoid(x):
    return 1 / (1 + exp(-x))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=241)
```

```python
for n in [1, 0.5, 0.3, 0.2, 0.1]:
    clf = GradientBoostingClassifier(n_estimators=250, verbose=True, random_state=241, learning_rate=n)
    clf.fit(X_train, y_train)
    ## calc. decision function on each step of training
    y_train_pred = clf.staged_decision_function(X_train)
    y_test_pred =  clf.staged_decision_function(X_test)
    
    test_loss = []
    train_loss = []
    ##get predicted values in convenient shape for log-loss function
    for i in y_test_pred :
        k = []
        for j in i:
            k.append(sigmoid(j[0]))
        test_loss.append(log_loss(y_test,k))
    for i in y_train_pred :
        k = []
        for j in i:
            k.append(sigmoid(j[0]))
        train_loss.append(log_loss(y_train,k))
    
    # for specific learning rate we need to calculate optimal number of steps for test data
    if n == 0.2:
        mn = 1000
        cur_i = 0
        for i in test_loss:
            if i < mn:
                mn = i
                cur_i = list(test_loss).index(i)
        print(cur_i, round(mn,2))
    # plot loss on test and train to understand how it goes through iterations
    plt.figure()
    plt.plot(test_loss, 'r', linewidth=2)
    plt.plot(train_loss, 'g', linewidth=2)
    plt.legend(['test', 'train'])
    plt.show()
   

    
```

#### fit random forest on best found params for GBC

```python
clf = RandomForestClassifier(random_state=241, n_estimators=36)
clf.fit(X_train, y_train)
```

```python
probs = clf.predict_proba(X_test)
#probs = [i[0] for i in probs]
round(log_loss(y_test,probs),2)
```
